```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# 1. Connect to the existing database
conn = sqlite3.connect('sales_data.db')

# 2. SQL query to get total quantity and total revenue per product
query = '''
SELECT 
    product, 
    SUM(quantity) AS total_quantity, 
    SUM(quantity * price) AS total_revenue 
FROM sales 
GROUP BY product
'''

# 3. Load query results into a pandas DataFrame
df = pd.read_sql_query(query, conn)
conn.close()

# 4. Print the result
print("ðŸ“Š Basic Sales Summary:")
print(df)

# 5. Plot a bar chart of revenue by product
df.plot(kind='bar', x='product', y='total_revenue', legend=False)
plt.title('Total Revenue by Product')
plt.ylabel('Revenue ($)')
plt.xlabel('Product')
plt.tight_layout()
plt.savefig("sales_chart.png")  # optional: saves the chart as PNG
plt.show()
```
